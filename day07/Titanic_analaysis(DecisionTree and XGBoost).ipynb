```python
from xgboost import  XGBClassifier as xgb
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv('test.csv')
```

```python
#Test passenger ids for later use
test_passenger_ids=test_df['PassengerId']
```

```python
#using a df called combined_df to process together
combined_df=pd.concat([train_df.drop('Survived',axis=1),test_df],axis=0)
```

```python
combined_df.isnull().sum()
```

## Preprocessing

```python
combined_df['Sex']=LabelEncoder().fit_transform(combined_df['Sex'])
```

```python
combined_df['Embarked']=combined_df['Embarked'].fillna('S')
combined_df['Embarked']=LabelEncoder().fit_transform(combined_df['Embarked'])
```

#### Simple Imputer for Age and Fare missing values

```python
from sklearn.impute import SimpleImputer
```

```python
numimpute=SimpleImputer(strategy='median')
```

```python
combined_df[['Age','Fare']]=numimpute.fit_transform(combined_df[['Age','Fare']])
```

## Feature Selection

```python
X=combined_df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
X_train=X.iloc[:len(train_df)]
```

```python
X_test=X.iloc[len(train_df):]
```

```python
y_train=train_df['Survived']
```

```python
model=xgb(eval_metric='logloss')
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
submission=pd.DataFrame({
    'passengerId': test_passenger_ids,
    'Survived': predictions
})
```

```python
submission.to_csv("submission.csv",index=False)
print("Submission.csv created successfully")
```
