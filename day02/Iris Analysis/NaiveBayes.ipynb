## Naive Bayes on Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
#Removed Iris-
df['Species'] = df['Species'].str.replace('Iris-', '', regex=False)
df
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by Species and compute the mean SepalLengthCm
grouped = df.groupby('Species').mean()

# Pivot to make it suitable for a heatmap
heatmap_data = grouped.pivot_table(index='Species', values='SepalLengthCm')

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", cbar=True)
plt.title("Average Sepal Length per Species")
plt.show()
```
