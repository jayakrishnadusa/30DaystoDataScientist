```python
"""We have used OnehotEncoder in the column transform example for the day02, Please refer for more https://ryanandmattdatascience.com/one-hot-encoder"""
```

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'], 'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df=pd.DataFrame(data=d)
```

```python
df.head(10)
```

```python
df['city'].unique()
```

```python
# If handle is unknown values from the dataset then we give it handle_unknown='ignore'
ohe=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ohetransform=ohe.fit_transform(df[['city']])
```

```python
ohetransform
```

```python
df=pd.concat([df,ohetransform],axis=1).drop(columns='city')
```

```python
df
```

```python
# Similarly for size
ohetransform1=ohe.fit_transform(df[['size']])
```

```python
ohetransform1
```

```python
df=pd.concat([df,ohetransform1],axis=1).drop(columns='size')
```
